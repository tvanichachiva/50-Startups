## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

## Import Data

```python
df = pd.read_csv(r'..\50_startups\50_Startups.csv')
df
```

## Get a unique list of states
 Looks like there are only 3 states in this data set, New York, California, and Florida

```python
# Check to see all unique values from the State field
states = df['State'].unique()
', '.join(map(str, states))
```

## Setting independent and dependent variables

```python
# Values from independent variables
X = df.iloc[:, :-1].values

# Values from depedent variable
y = df.iloc[:, -1].values
```

```python
# OneHotEncode categorical variable (state)
st = ColumnTransformer([("State", OneHotEncoder(), [3])],
                       remainder='passthrough')
X = st.fit_transform(X)

# Remove one column to avoid dummy variable trap, since there are only 3 states
X = X[:, 1:]
```

```python
X
```

## Create training and testing data sets and load the linear regression model

```python
# Create train/test data sets
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.25,
                                                    random_state=0)

# Assign LinearRegression class to reg object
reg = LinearRegression()

# Fit model to training data
reg.fit(X_train, y_train)
```

## Predict values on the test set using trained model

```python
# Prediction
y_pred = reg.predict(X_test)

# R Squared score
print(f'The R-Squared Score is: {round((r2_score(y_test, y_pred)),4) * 100}%')
```

## Use Backwards Elimination to tune the model
Without eliminating any indepedent variables our r-squared score was 93%. Using backwards elimination we will see if the r-squared score can be improved

```python
# Add an array of ones to the X dataset (independent variables) and call it X_opt
# X_opt will be our optimal set of features
num_obs = len(X)
X = np.append(arr=np.ones((num_obs, 1)).astype(float), values=X, axis=1)
X_opt = np.array(X[:, :], dtype=float)
```

```python
# Using the OLS algorithim, we will iteratively remove features that have a p-value greater than 0.05
# Will need to remove one feature at a time
model = sm.OLS(endog=y, exog=X_opt)
reg_OLS = model.fit()
reg_OLS.summary()
```

## First Feature to remove
- Dummy Variable for state
- ~~Dummy Variable for state~~

- Admin Spend
- Marketing Spend

x2 (Second hot encoded variable) will need to be removed. It's p-value is above 0.05 and is the highest

```python
# Delete x2 (3rd index/Dummy Variable for state)
X_opt = np.delete(X_opt, 2, axis=1)

# Rerun OLS algorithim
model = sm.OLS(endog=y, exog=X_opt)
reg_OLS = model.fit()
reg_OLS.summary()
```
